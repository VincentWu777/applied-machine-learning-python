# Problem Formulation

Before building any machine learning model, we must clearly define
the problem, the inputs, the outputs, and the success criteria.


## 1. Problem Description

The goal of this notebook is to understand how to translate a real-world
question into a machine learning problem.

At this stage, no model is built and no data preprocessing is performed.


## 2. Is Machine Learning Appropriate?

Machine learning is appropriate when:
- The relationship between inputs and outputs is complex or unknown
- Explicit rules are difficult to define
- Sufficient data is available

If simple rules or deterministic logic can solve the problem,
machine learning should not be used.


## 3. Learning Setup

This project focuses on **supervised learning**.

- **Input (X):** A set of measurable features describing each example
- **Output (y):** A target variable to be predicted
- **Task type:** Classification or regression depending on the problem


## 4. Success Criteria

Model performance should be evaluated using appropriate metrics,
not just accuracy.

The chosen metric must reflect the real-world cost of errors
(e.g., false positives vs false negatives).


## 5. Common Pitfalls

- Choosing a model before understanding the data
- Evaluating on training data
- Ignoring class imbalance
- Optimizing metrics without understanding their meaning
